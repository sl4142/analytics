```python
maxCorr, mkt = 0, ""
for key,data in data_dict.items():
    corr = cf.get_correlation(data_dict['BTC-USD']['Adj Close'], data['Adj Close'])
    if cf.isNan(corr) or corr == 1:
        continue
    maxCorr = max(abs(corr), maxCorr)
#     print(key, corr, maxCorr)
    if maxCorr == abs(corr):
        mkt = key
print(maxCorr, mkt)
```
