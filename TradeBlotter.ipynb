```python
def addTrade(tradeBlotter, entryDate, entryPrice, direction, exitPrice, live, market, 
             realizedPnl, unrealizedPnl, stopLevel, exitDate, stopOrExit):
    tradeBlotter.entryDate.append(entryDate.strftime('%Y-%m-%d'))
    tradeBlotter.entryPrice.append(entryPrice)
    tradeBlotter.direction.append(direction)
    tradeBlotter.exitPrice.append(exitPrice)
    tradeBlotter.live.append(live)
    tradeBlotter.market.append(market)
    tradeBlotter.realizedPnl.append(realizedPnl)
    tradeBlotter.unrealizedPnl.append(unrealizedPnl)
    tradeBlotter.stopLevel.append(stopLevel)
    tradeBlotter.exitDate.append(exitDate)
    tradeBlotter.stopOrExit.append(stopOrExit)
    tid = len(tradeBlotter.entryPrice)-1
    if not market in tradeBlotter.openTrades:
        tradeBlotter.openTrades[market] = [tid]
    else:
        tradeBlotter.openTrades[market].append(tid)
    return tid

def getUnrealizedPnl(tradeBlotter, curPrice, market):
    unrealizedPnl = 0
    if not market in tradeBlotter.openTrades:
        return unrealizedPnl
    for tid in tradeBlotter.openTrades[market]:
        tradeBlotter.unrealizedPnl[tid] = (curPrice - tradeBlotter.entryPrice[tid]) * tradeBlotter.direction[tid]
        unrealizedPnl += tradeBlotter.unrealizedPnl[tid]
    return unrealizedPnl
        
        
def closeTrade(tradeBlotter, curPrice, curDate, market):
    realizedPnl = 0
    while tradeBlotter.openTrades[market]:
        tid = tradeBlotter.openTrades[market][0]
        tradeBlotter.openTrades[market] = tradeBlotter.openTrades[market][1:]  # pop the first tid
        tradeBlotter.exitPrice[tid] = curPrice
        tradeBlotter.exitDate[tid] = curDate.strftime('%Y-%m-%d')
        tradeBlotter.unrealizedPnl[tid] = 0
        tradeBlotter.live[tid] = 0
        tradeBlotter.realizedPnl[tid] = (curPrice - tradeBlotter.entryPrice[tid]) *  tradeBlotter.direction[tid]
        realizedPnl += tradeBlotter.realizedPnl[tid]
    del tradeBlotter.openTrades[market]
    return realizedPnl
    
def updatePortfolio(tradeBlotter, mkt, entryPrice, entryDate, unit):
    entryDate = entryDate.strftime('%Y-%m-%d')
    portfolio = {} if tradeBlotter.lastPortfolioUpdateDate == None else tradeBlotter.portfolioHistory[tradeBlotter.lastPortfolioUpdateDate]   # problem here. should carry over the last portfolio
    details = Details(unit, entryPrice, entryDate)
    if mkt in portfolio:
        portfolio[mkt].append(details)
    else:
        portfolio[mkt] = [details]
    tradeBlotter.portfolioHistory[entryDate] = portfolio
    tradeBlotter.lastPortfolioUpdateDate = entryDate
```

```python
def backtest(mktData, tradeBlotter, startDate, endDate, totalCapital):
    dateList = cf.get_dates_list(startDate, endDate)
    navs = [totalCapital]*len(dateList)
    cash = totalCapital
    for i, curDate in enumerate(dateList):
        realizedPnl, unrealizedPnl, curDate = 0, 0, pd.Timestamp(curDate.year, curDate.month, curDate.day)
        for mkt, data in mktData.items():
            curPrice = data['Adj Close'][curDate]
            if not curDate in data.index:
                continue
            if data['Adj Close'][curDate] > 10000 and cash >= curPrice:    # CHECK BUY
                cash -= curPrice
                tid = addTrade(tradeBlotter, curDate, curPrice, 1, np.nan, 1, mkt, np.nan, np.nan, curPrice*0.7, np.nan, np.nan)
                updatePortfolio(tradeBlotter, mkt, curPrice, curDate, 1)
            if data['Adj Close'][curDate] < 10000 and mkt in tradeBlotter.openTrades: # CHECK SELL
                realizedPnl += closeTrade(tradeBlotter, curPrice, curDate, mkt)
                cash += realizedPnl
            unrealizedPnl += getUnrealizedPnl(tradeBlotter, curPrice, mkt)
        if i != 0:
            if (i == len(dateList)-6):
                print(navs[i-1], realizedPnl, unrealizedPnl)
            navs[i] = navs[i-1] + realizedPnl + unrealizedPnl
    return pd.DataFrame(navs, index=pd.Series(dateList).apply(lambda x: x.strftime('%Y-%m-%d')), columns=['NAV'])
                    
```
